```python
import hvplot.pandas  # noqa
import pandas as pd

soil_moisture_df = pd.read_csv(
    "~/drysat_jupyterhub/ascat_insitu_merge/obs_data_long.csv", index_col="datetime", parse_dates=True
)

soil_moisture_df
```

```python


soil_moisture_df.hvplot.line(
    x="datetime",
    y="value",
    row="location",
    col="type",
    frame_height=200,
    frame_width=250,
)

density_df = pd.DataFrame(
    {
        "location": ["Buzi", "Chokwé", "Mabalane", "Mabote", "Muanza"],
        "bulk density": [1.25, 1.4, 1.4, 1.35, 1.25],
    }
).set_index("location")
density_df

soil_moisture_df = soil_moisture_df.merge(
    density_df, left_on="location", right_index=True
)
soil_moisture_df

def rescale(x):
    if x.type == "ascat":
        value, density = x.value, x["bulk density"]
        porosity = 1 - density / 2.65
        return porosity * value / 100
    else:
        return x.value


soil_moisture_df["value"] = soil_moisture_df.apply(rescale, axis=1)
```

```python

soil_moisture_df

soil_moisture_df.hvplot.line(
    x="datetime",
    y="value",
    row="location",
    col="type",
    frame_height=200,
    frame_width=250,
)

```
